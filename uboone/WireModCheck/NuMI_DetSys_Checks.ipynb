```python
def get_different_rows(source_df, new_df):
    """Returns just the rows from the new dataframe that differ from the source dataframe"""
    merged_df = source_df.merge(new_df, indicator=True, how='outer')
    changed_rows_df = merged_df[merged_df['_merge'] == 'right_only']
    return changed_rows_df.drop('_merge', axis=1)
```

```python
for ftuple in roi_files:
    print("Processing ",ftuple[0]," and ",ftuple[1])
    df_v48_rois = uproot.open(ftuple[0])['wireana/roi_tree'].pandas.df(flatten=False)
    df_crb_rois = uproot.open(ftuple[1])['wireana/roi_tree'].pandas.df(flatten=False)
    
    counter=0
    
    for i, row in df_v48_rois.iterrows():
        row_crb = df_crb_rois.loc[i]
        #if not row.equals(row_crb):
        if row['roi_start_tick']!=row_crb['roi_start_tick']:
            print("Difference start tick!!")
            print(row,row_crb)
            print("\n")
        elif not np.isclose(row['waveform'],row_crb['waveform'],rtol=0.00001).all():
            print("Difference waveforms!!")
            print((row_crb['waveform']-row['waveform'])/row['waveform'])
            print("\n")
        
        if counter%1000==0:
            print("Counter %d\n"%counter,
                  row,
                  "\n",
                  row_crb,
                  "\n",
                  (row_crb['waveform']-row['waveform'])/row['waveform'])
        counter+=1
            
```
